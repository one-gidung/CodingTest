s	   answer\
"()()"	true\
"(())()"	true\
")()("	false\
"(()("	false

```python
from collections import deque

def solution(s):
    tmp = deque()

    #   ")"  False
    # stack "(" insert ")"  "(" pop  
    # stack  ....
    
    # insert append
    # popright pop
    
    for i in range(len(s)):
        
        if s[i] == "(":
            tmp.append(s[i])
        else:         # s[i] == ")"
            if i == 0 or len(tmp)==0:
                return False
            tmp.pop()
    if len(tmp) != 0:
        return False

    return True
```
