입출력 예\
\
s	   answer\
"()()"	true\
"(())()"	true\
")()("	false\
"(()("	false

```python
from collections import deque

def solution(s):
    tmp = deque()

    # 첫번째 괄호가 ")"이면 무조건 False
    # stack에 "(" insert하고 ")"를 만나면 "("를 pop해버리면 될 듯
    # stack을 어떻게 구현하더라....흠
    
    # insert는 append
    # popright는 pop
    
    for i in range(len(s)):
        
        if s[i] == "(":
            tmp.append(s[i])
        else:         # s[i] == ")"
            if i == 0 or len(tmp)==0:
                return False
            tmp.pop()
    if len(tmp) != 0:
        return False

    return True
```

```python
solution("(()(")
```
