## sort와 sorted의 차이
- sort는 리스트 자체를 정렬시켜버린다.
- sort는 반환값이 없다 (None)



- sorted는 정렬된 리스트를 반환값으로 준다.
- sorted는 원래 리스트를 정렬하지 않는다.
- 정렬된 새로운 리스트를 반환값으로 준다 (list)
