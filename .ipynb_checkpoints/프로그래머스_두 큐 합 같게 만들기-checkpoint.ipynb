# 리스트로 풀기 (시간초과 뜸)

```python
def solution(q1, q2):
    
    # q1과 q2의 합을 비교
    
    # 같으면 stop
    
    # 다르면 큰거에서 popleft해서 작은 쪽에 insert
    
    # 그러다가 cnt가 전체 q1, q2 원소 갯수의 합보다 커지면 stop = 불가능
    
    # 이때 pop하는 쪽을 하나로 정해두고 실행 후 끝나면 다른쪽도 실행해야함
    
    cut = (len(q1)+len(q2))*2
    cnt_list = []
    count = 0
    while count <= cut:

        # q1 q2 합 비교
        if sum(q1) > sum(q2):
            q2.append(q1.pop(0)) # q1에서 index가 0인 거 pop해서 q2에 삽입
            count += 1
        elif sum(q1) < sum(q2):
            q1.append(q2.pop(0)) # q2에서 index가 0인 거 pop해서 q1에 삽입
            count += 1
        else: # sum(q1)==sum(q2)
            cnt_list.append(count) # list에 정답 횟수 저장
            count = 0 # count 0으로 초기화
            break

    if len(cnt_list) == 0:
        ans = -1
    else:
        ans = min(cnt_list)

    

    return ans
```

```python
solution([2,1,1,3,1,7], [1])
```

# 큐로 풀기

```python
from collections import deque

def solution(queue1, queue2):
    # 디큐 만들기
    q1 = deque(queue1)
    q2 = deque(queue2)
#     q.popleft() # 왼쪽 원소 추출
#     q.append() # 오른쪽에 원소 삽입

    cut = (len(q1)+len(q2))*2
    cnt_list = []
    count = 0
    s1, s2 = sum(q1), sum(q2)
    
    while count <= cut:

        # q1 q2 합 비교
        if s1 > s2:
            s1 -= q1[0]
            s2 += q1[0]
            q2.append(q1.popleft()) # q1에서 pop해서 q2에 삽입
            count += 1
        elif s1 < s2:
            s1 += q2[0]
            s2 -= q2[0]
            q1.append(q2.popleft()) # q2에서 pop해서 q1에 삽입
            count += 1
        else: # s1==s2
            cnt_list.append(count) # list에 정답 횟수 저장
            count = 0 # count 0으로 초기화
            break

    if len(cnt_list) == 0:
        ans = -1
    else:
        ans = min(cnt_list)

    return ans
```

```python
solution([2,1,1,3,1,7], [1])
```
