#   ( )

```python
def solution(q1, q2):
    
    # q1 q2  
    
    #  stop
    
    #   popleft   insert
    
    #  cnt  q1, q2     stop = 
    
    #  pop        
    
    cut = (len(q1)+len(q2))*2
    cnt_list = []
    count = 0
    while count <= cut:

        # q1 q2  
        if sum(q1) > sum(q2):
            q2.append(q1.pop(0)) # q1 index 0  pop q2 
            count += 1
        elif sum(q1) < sum(q2):
            q1.append(q2.pop(0)) # q2 index 0  pop q1 
            count += 1
        else: # sum(q1)==sum(q2)
            cnt_list.append(count) # list   
            count = 0 # count 0 
            break

    if len(cnt_list) == 0:
        ans = -1
    else:
        ans = min(cnt_list)

    

    return ans
```

```python
from collections import deque

def solution(queue1, queue2):
    #  
    q1 = deque(queue1)
    q2 = deque(queue2)
#     q.popleft() #   
#     q.append() #   

    cut = (len(q1)+len(q2))*2
    cnt_list = []
    count = 0
    s1, s2 = sum(q1), sum(q2)
    
    while count <= cut:

        # q1 q2  
        if s1 > s2:
            s1 -= q1[0]
            s2 += q1[0]
            q2.append(q1.popleft()) # q1 pop q2 
            count += 1
        elif s1 < s2:
            s1 += q2[0]
            s2 -= q2[0]
            q1.append(q2.popleft()) # q2 pop q1 
            count += 1
        else: # s1==s2
            cnt_list.append(count) # list   
            count = 0 # count 0 
            break

    if len(cnt_list) == 0:
        ans = -1
    else:
        ans = min(cnt_list)

    return ans
```

```python
solution([2,1,1,3,1,7], [1])
```
